### Which Industry Pays the most

```python
industry_salary = (
    salary_survey
        .groupby('Industry')['Annual Salary']
        .agg(mean_salary='mean',
             median_salary='median',
             count='count')
        .sort_values('mean_salary', ascending=False)
    .head(10)
)

industry_salary
```

```python
plt.figure(figsize=(10,6))

industry_salary['mean_salary'].plot(
    kind='bar'
)

plt.title('Top 10 Industries by Average Salary')
plt.xlabel('Industry')
plt.ylabel('Average Annual Salary')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Salary growth with years of experiece

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=salary_survey,
    x='Overall Work Experience',
    y='Annual Salary'
)

plt.xticks(rotation=45)
plt.title('Salary Distribution by Overall Work Experience')
plt.xlabel('Overall Work Experience')
plt.ylabel('Annual Salary')
plt.tight_layout()
plt.show()
```
